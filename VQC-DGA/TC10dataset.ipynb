```python
# ==============================================================
# 0. Thư viện cần thiết
# ==============================================================
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

from qiskit_aer import AerSimulator
from qiskit.circuit.library import EfficientSU2
from qiskit_algorithms.optimizers import COBYLA
from qiskit_machine_learning.algorithms import VQC
from qiskit.visualization import circuit_drawer
from qiskit.circuit import ParameterVector, QuantumCircuit

# ==============================================================
# 1. Đọc dữ liệu & chia 588/117
# ==============================================================
file_path = "X.xlsx"          # Đường dẫn file
df        = pd.read_excel(file_path)

TRAIN_SIZE = 587
TEST_SIZE  = 118
assert len(df) >= TRAIN_SIZE + TEST_SIZE, "Dataset không đủ 705 mẫu!"

# Cột 0 = nhãn, cột 1-4 = đặc trưng
y_full = df.iloc[:, 0].values
X_full = df.iloc[:, 1:5].values          # (n_samples, 4)

# ==============================================================
# 2. Mã hoá nhãn & CHUẨN HOÁ TRÊN TOÀN BỘ DỮ LIỆU
#    (sau đó mới cắt train / test)
# ==============================================================
le = LabelEncoder()
y_full_enc = le.fit_transform(y_full)

scaler = StandardScaler()
X_full_scaled = scaler.fit_transform(X_full)     # ⬅️ fit trên toàn bộ 705 mẫu

# Chia thủ công sau khi đã chuẩn hoá
X_train = X_full_scaled[:TRAIN_SIZE]
X_test  = X_full_scaled[TRAIN_SIZE:TRAIN_SIZE+TEST_SIZE]

y_train = y_full_enc[:TRAIN_SIZE]
y_test  = y_full_enc[TRAIN_SIZE:TRAIN_SIZE+TEST_SIZE]


# ==============================================================
# 3. Feature map tham số hoá & Ansatz
# ==============================================================
def parameterized_feature_map(num_qubits=2, feature_dimension=4):
    params = ParameterVector('x', feature_dimension)
    qc = QuantumCircuit(num_qubits)
    qc.ryy(params[0], 0, 1)
    qc.rzx(params[2], 1, 0)
    qc.ryy(params[1], 1, 0)
    qc.rzx(params[3], 0, 1)
    return qc

num_qubits  = 2
feature_map = parameterized_feature_map(num_qubits, feature_dimension=4)
ansatz      = EfficientSU2(num_qubits, entanglement= 'sca', reps=4)

optimizer   = COBYLA(maxiter=300)
backend     = AerSimulator()

# (tuỳ chọn) Hiển thị mạch
# ----------------------------------------
print("🔹 Mạch Feature Map:")
display(circuit_drawer(feature_map.decompose(), output='mpl', style="clifford", fold=20))
plt.show()

print("🔹 Mạch Ansatz:")
display(circuit_drawer(ansatz.decompose(), output='mpl', style="clifford", fold=20))
plt.show()

# ==============================================================
# 4. Huấn luyện VQC One-Vs-Rest
# ==============================================================
classifiers = {}
print("🔹 Huấn luyện VQC (One-Vs-Rest)...")
for label in np.unique(y_train):
    y_bin = (y_train == label).astype(int)      # nhãn nhị phân
    vqc   = VQC(optimizer=optimizer,
                feature_map=feature_map,
                ansatz=ansatz)
    vqc.fit(X_train, y_bin)
    classifiers[label] = vqc
    print(f"   • Đã huấn luyện lớp {label}")

# ==============================================================
# 5. Hàm dự đoán & đánh giá
# ==============================================================
def predict_ovr(X):
    preds, all_probs = [], []
    for idx, x in enumerate(X):
        x_ = x.reshape(1, -1)
        probs = {lbl: clf.predict_proba(x_)[0][1]
                 for lbl, clf in classifiers.items()}
        all_probs.append(probs)
        print(f"Mẫu {idx+1}: " +
              ", ".join([f"L{lbl}: {p:.3f}" for lbl, p in probs.items()]))
        preds.append(max(probs, key=probs.get))
    return np.array(preds), all_probs

y_pred, _ = predict_ovr(X_test)

print("\n🔹 Kết quả đánh giá:")
print(f"   Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(f"   Precision: {precision_score(y_test, y_pred, average='weighted'):.4f}")
print(f"   Recall   : {recall_score(y_test, y_pred, average='weighted'):.4f}")
print(f"   F1-score : {f1_score(y_test, y_pred, average='weighted'):.4f}")

print("\n🔹 So sánh từng mẫu test:")
for i in range(len(y_test)):
    real = le.inverse_transform([y_test[i]])[0]
    pred = le.inverse_transform([y_pred[i]])[0]
    print(f"   Mẫu {i+1:3d}:  Dự đoán = {pred:<10} | Thực tế = {real}")
```
